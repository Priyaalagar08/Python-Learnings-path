**List**

```python
# List
# can be modified
# Ordered
# can be duplicated
print([1,2,3,4]) # square brackets
```

```python
# Adding one value in the list
my_list = [5,'Priya','3.6',6.7e307]
my_list.append(10)
print(my_list)
```

```python
# Changing one element in the list
my_list[1] = 'Ramya'
print(my_list)
```

**Tuple**

```python
# Tuple
# Can't be modified
# Ordered
# can be duplicated
print((1,1,2,3,4)) # parantheses
```

```python
# Trying to modify the tuple values
my_tuple = (1,2,3,4,5)
my_tuple[1] = 10
```

**Sets**

```python
# Sets
# can be modified
# Unordered
# Can't be duplicated- only unique values
print({0,1,2,3,4,5}) # curly brackets
```

```python
# Adding one value in the set
my_set = {5,'Priya','3.6',6.7e307}
my_set.add(10)
print(my_set)
# We can see the output is not in order
```

**Dictionary**

```python
# Dictionary
# can be modified
# Unordered
# Key value can be duplicated
# Pair of values are stored
# string, integer & tuple can be used as a key since they are Immutable
print({'Name':'Priya', 'Age':27,'Gender':'Female'}) # curly brackets
```

```python
my_dict = {'Name':'Priya', 'Age':27,'Gender':'Female'}
print(my_dict)
```

```python
# Adding new key pair
my_dict['job'] = "Engineer"
print(my_dict)
```

```python
# Changing the values in the keys
my_dict['Name'] = 'Ramya'
print(my_dict)
```

```python
# changing the key name
my_dict['gender'] = my_dict['Gender']
print(my_dict)
```

**Keywords &** **Identifier**

```python
# We write program in english which is converted to binary language by the compiler and interpreter to make machine understandable so there are about 32 keywords in python
# whatever name we create in a programming say variable, functions all are identifiers
#keywords are defined by python eg: if, else, while, for, def, class, True, false so we have almost 32 keywords which can't be used as variables name
# identifiers are naming elements (variable identifiers, function identifiers, class identifiers)

# We cannot start variable with a digit like 2name = 'priya'(gives error) but we can use name2='priya'
name5 = 'Priya'
print(name5)
# we can use only one special characters which is _ underscore remaining characters gives error
_ = 'Priya'
print(_)
name_name = 'Priya'
print(name_name)
# Identifiers cannot use keywords
```
