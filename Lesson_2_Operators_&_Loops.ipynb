##### There are 6 operators in python
##### 1. Arithmetic Operators + - * /
##### 2.Relational Operators > < ==
##### 3.Logical Operators AND NOT OR
##### 4.Bitwise Operators AND OR XOR NOT right and left shift
##### 5.Assignment Operators =
##### 6.Membership Operators In , Not in


```python
# Arithmatic Operators
a = 8
b = 2
print (a + b)
print (a - b)
print (a * b)
print (a / b)
print (a // b) ## Interger devision
print(a % b) ## gives reminder
print(a**b) #a to the power b
```

```python
# Relational Operators
print (5 > 10)
print (5 < 10)
print (5 == 5)
print (5 >= 10)
print (5 <= 10)
print (5!= 5)
```

```python
# Logical Operators
print (1 and 0)
print (0 and 0)
print (1 and 1)
print (1 or 0)
print (not 0)
```

```python
# Bitwise Operators

# bitwise and
print (5 & 2)

# bitwise or
print (5 | 2)

# bitwise xor
print (5 ^ 2)

# bitwise not
print (~5)

# bitwise left shift
print (5 << 2)

# bitwise right shift
print (5 >> 2)
```

```python
# Assignment operator (=)
a = 10
b = 5
c = 8
d = 6
a +=2
print (a)
b -=2
print (b)
c *=2
print (c)
d /=2
print (d)
```

```python
# Membership Operator in/not in this operators works on all datatypes like int, string, list, tuple and dictionary

print ('C' not in 'chennai')
print (1 in (1,2,3))
print (4 in [1,2,3])
print (3 in {1,2,3})
```

# **If Else and Nested If Else**
 ##### If only two conditions need to be executed then if else is sufficient. If more than two conditions need to be performed then elif is used. If else is again used under elif which is called nested if else.

```python
# Use if else code to show the entered email is correct or not
# email--> abc@gmail.com
# password --> 1234
email = input('Enter the email address:')
password = input('Enter the password')
if email=='abc@gmail.com' and password=='1234':
  print('Login successful')
elif email=='abc@gmail.com' and password!='1234':
  print('Password is incorrect')
  password=input('Enter the correct password:')
  if password=='1234': # nested if else
    print('Login successful')
  else:
    print('Login failed')
else:
  print('Not correct')
```

```python
# Use if else to print smallest number. Get 3 numbers from user
a = int(input('Enter 1st num:'))
b = int(input('Enter 2nd num:'))
c = int(input('Enter 3rd num:'))
if a<b and a<c:
  print(a, 'is smallest')
elif b<a and b<c:
  print(b, 'is smallest')
elif a==b==c:
  print('All are equal')
else:
  print(c,'is smallest')
```

```python
# Menu driven calculator
a = int(input('Enter 1st num:'))
b = int(input('Enter 2nd num:'))
Operation = input(("Enter the Operation:"))

if Operation == '+':
  print(a+b)
elif Operation == '-':
  print(a-b)
elif Operation == '*':
  print(a*b)
else:
  print(a/b)
```

**Loops in Python**

While Loop

```python
# Tables program
number = int(input('Enter the number'))
i = 1
while i<11:
  print(number,'*',i,'=',number * i)
  i+=1
```

```python
# while loop with else
x = 1
while x<4:
  print(x)
  x+=1
else:
  print('Limit Crossed')
```

```python
# Guessing game
# generate a random integer beween 1 to 100
import random
jackpot= random.randint(1,100)

Guess= int(input('Guess the number'))
count = 1
while Guess!=jackpot:
  if Guess < jackpot:
    print('Wrong, guess higher')
  elif Guess > jackpot:
    print('Wrong, guess lower')

  Guess= int(input('Guess the number'))
  count+=1

else:
  print('Correct')

  print('attemps:',count)



```

**For Loop**

```python
# for loop iteration in list
for i in [1,2,3]:
  print(i)
```
